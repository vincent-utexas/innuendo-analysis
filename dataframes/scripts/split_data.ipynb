```python
import pandas as pd
from sklearn.model_selection import train_test_split

# comments, unused_comments = train_test_split(comments, test_size=0.5, stratify=comments['label'], random_state=42)

mlm_df, classification_df = train_test_split(comments, test_size=0.5, stratify=comments['label'], random_state=42)
```

```python
# for mlm_df, only care about class 1

mlm_df, extra_df = mlm_df[mlm_df['label'] == 1], mlm_df[mlm_df['label'] == 0]
```

```python
mlm_df.to_csv('/content/drive/My Drive/LIN371/transfer_learning/mlm_df.csv', index=False)
extra_df.to_csv('/content/drive/My Drive/LIN371/transfer_learning/extra_df.csv', index=False)
classification_df.to_csv('/content/drive/My Drive/LIN371/transfer_learning/classification_df.csv', index=False)
```

```python
# further split classification into train val test

classification_df, unused_comments = train_test_split(classification_df, test_size=0.5, stratify=classification_df['label'], random_state=42)

train_df, test_df = train_test_split(classification_df, test_size=0.5, stratify=classification_df['label'], random_state=42)
val_df, test_df = train_test_split(test_df, test_size=0.5, stratify=test_df['label'], random_state=42)
```

```python
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
```

```python
mlm_df = mlm_df.reset_index(drop=True)
extra_df = extra_df.reset_index(drop=True)
```

```python
train_df.to_csv('/content/drive/My Drive/LIN371/train_df.csv', index=False)
val_df.to_csv('/content/drive/My Drive/LIN371/val_df.csv', index=False)
test_df.to_csv('/content/drive/My Drive/LIN371/test_df.csv', index=False)
mlm_df.to_csv('/content/drive/My Drive/LIN371/mlm_df.csv', index=False)
extra_df.to_csv('/content/drive/My Drive/LIN371/extra_df.csv', index=False)
```
