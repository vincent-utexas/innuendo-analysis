```python
import re

def remove_non_alpha_except_space(input_string):
    # Replace all characters that are not alphabetic or spaces with an empty string
    result = re.sub(r'[^a-zA-Z\s]', ' ', input_string)
    return result.lower()

# Example usage
input_string = "Hello, World! 123. How's it going?"
cleaned_string = remove_non_alpha_except_space(input_string)

print(cleaned_string)  # Output: Hello World Hows it going
```

```python
comments = pd.read_csv('/content/drive/My Drive/LIN371/sexually_explicit_text_downsampled.csv')
comments.head()
```

```python
explicit_words = result['text'].tolist()
# print(explicit_words[:5])

len1words = [word for word in explicit_words if len(word.split()) == 1]
longerwords = [word for word in explicit_words if len(word.split()) > 1]
print(longerwords[:5])
```

```python
def classify(body, label):
  has_explicit_words = False
  text = remove_non_alpha_except_space(body)
  for word in longerwords:
    if word in text :
      has_explicit_words = True
      break
  for word in len1words:
    if len(word) > 4:
      if word in text:
        has_explicit_words = True
        break
  text = text.split()
  for word in len1words:
    if word in text:
      has_explicit_words = True
      break

  if has_explicit_words == True and label == 1:
    return 'explicit_source_has_explicit_words'

  elif has_explicit_words == False and label == 1:
    return 'explicit_source_no_explicit_words'

  elif has_explicit_words == True and label == 0:
    return 'control_source_has_explicit_words'

  elif has_explicit_words == False and label == 0:
    return 'control_source_no_explicit_words'

```

```python
comments['new_label'] = comments.apply(lambda row: classify(row['body'], row['label']), axis=1)
```
